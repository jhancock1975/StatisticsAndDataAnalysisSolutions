<h2>Question 2</h2>
<p>Question 2 asks for comparison dot plots.  In the text above these questions in \cite{stackSeabornXAxis} the authors define a comparison dot plot as a display of dot plots that are vertically stacked.  We create graphics conveying the same data using the Seaborn countplot function.   The comparison plots that we must generate to answer this question show batting averages by hitter type.  The comparison plots are below.  However, first we 
must read in the batting average data and partition it into sets for right-handed, left-handed and switch hitter types.</p>

```python
bb_data = pd.read_csv('../baseball-data.csv', sep=' ', header=None)
# the sixth column of the baseball data is the batting average for hitters
# the baseball data comes with the convention that the row is a record of
# data for a hitter if the fifth column is a 1.
bat_avg = bb_data[bb_data[4]==1]
left_bat_avg = bat_avg[bat_avg[2]==2]
right_bat_avg = bat_avg[bat_avg[2]==1]
switch_bat_avg = bat_avg[bat_avg[2]==3]
```

Now we have the data prepared, we can generate the plots using a helper function with a for loop.

```python
def draw_countplot(data_to_plot, title):
    ax=sns.countplot(data_to_plot)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```

```python
for data_to_plot, title in zip([left_bat_avg[5], right_bat_avg[5], switch_bat_avg[5]], 
                              ['Left-handed batting averages', 'Right-handed batting averages', 
                              'Switch-hitter batting averages']):
    draw_countplot(data_to_plot, title)
```
