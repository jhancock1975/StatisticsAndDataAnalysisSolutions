<h2>Question 1</h2>
<p>The first part of question one asks us what is a typical value for weight in the baseball data. 
Since we just read through a discussion on stem-leaf plots, this is probably an opportunity to make a
stem-leaf plot in the same fashion that the authors write about above this question. 

<p>We found the information in \cite{pyStemLeaf} to generate the stem-leaf plot.

```python
bb_data = pd.read_csv('../baseball-data.csv', sep=' ', header=None)
```

```python
fig, ax = stemgraphic.stem_graphic(bb_data[1], break_on=5)
```

<p>So, a typical weight is between 185 and 195.</p>
The next part of the question is asking us if more baseball players weigh over 200 pounds than under 170 pounds.  From the stem-leaf plot above, we can see that there is only one row with a weight of 170 in it, but there are many more rows with weight values over 200, so our answer is yes, there are more baseball players that weight over 200 pounds.</p>
