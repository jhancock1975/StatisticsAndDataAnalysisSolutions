Not surprisingly, StackOverflow has an example that looks closer to what is in the book \cite{stackDotPlot}.  This is as close as I could get following the StackOverflow example, I am going to leave it at this for now, and use histograms for these dot plots since I believe they carry the same information.

Later we went back to the Kaggle data visualization tutorial on Plotly \cite{kagglePlotly} 

```python
import numpy as np; np.random.seed(13)
import matplotlib.pyplot as plt

bins_max=max(etruscan_head_sizes)
bins_min=min(etruscan_head_sizes)
print('bins_min: %d, bins_max %d' % (bins_min, bins_max))
bins = np.arange( bins_max)-0.5

hist, edges = np.histogram(etruscan_head_sizes, bins=bins)

y = np.arange(1,hist.max()+1)
x = np.arange(bins_max-1)
X,Y = np.meshgrid(x,y)

plt.scatter(X,Y, c=Y<=hist, cmap="Greys")

plt.show()
```

We used StackOverflow user ImportanceOfBeingErnest's answer from \cite{stackSeabornXAxis} for the formatting of the bar chart below.

```python
import matplotlib.pyplot as plt
import seaborn as sns
ax=sns.countplot(etruscan_head_sizes)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
